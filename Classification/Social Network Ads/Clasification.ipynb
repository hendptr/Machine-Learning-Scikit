```python
X = df.iloc[:, [2,3]].values
y = df.iloc[:, -1].values
```

#### Spliting Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

#### Feature Scalling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### Classification

##### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
y_pred = log_reg.predict(X_test)
score_log_reg = round(log_reg.score(X_test, y_test) * 100, 2)
print('Score : ', score_log_reg)
print('Confusion Matix \n', confusion_matrix(y_test, y_pred))
```

##### Support Vector Machine

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
score_svc = round(svc.score(X_test, y_test) * 100, 2)
print('Score : ', score_svc)
print('Confusion Matix \n', confusion_matrix(y_test, y_pred))
```

##### Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)
score_rfc = round(rfc.score(X_test, y_test) * 100, 2)
print('Score : ', score_rfc)
print('Confusion Matrix \n', confusion_matrix(y_test, y_pred))
```

#### Kesimpulan

```python
cons = pd.DataFrame({'Mode':['Logistic Regression', 'SVM', 'Random Forest'], 'Score':[score_log_reg, score_svc, score_rfc]})
```
