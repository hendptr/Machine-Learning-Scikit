```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('50_Startups.csv')
print(df.shape)
print(df.info())
print(df.describe())
```

```python
#Cek Korelasi Antar Variable
sns.heatmap(df.corr(), annot=True)

#Didapat Variabel R&D dan Marketing yang sangat berpengaruh terhadap profit
```

```python
X = df.iloc[:, [0, 2]].values
y = df.iloc[:, -1].values
```

```python
#Splitting Data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#Gunakan 2 Variable R&D dan Marketing ke dalam Model pada Training Set
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
linreg = LinearRegression().fit(X_train, y_train)
print('Linear Regression Score : ', linreg.score(X_test, y_test))
y_pred = linreg.predict(X_train)
print('Mean Squared Error :', mean_squared_error(y_pred, y_train))
print('Intercept : ', linreg.intercept_)
print('Coef : ', linreg.coef_)
d = pd.DataFrame({'y_pred':y_pred, 'y':y_train})
d
```

```python
#Prediksi Test Set
y_pred_test = linreg.predict(X_test)
print('Mean Squared Error : ', mean_squared_error(y_pred_test, y_test))
d = pd.DataFrame({'y_pred':y_pred_test, 'y':y_test})
d
```
